```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

dataset = sns.load_dataset("titanic")
sns.countplot(data=dataset, x='survived', hue='sex')
plt.show()
```

```python
dataset['sex'].value_counts().plot(kind = "pie", autopct = "%.2f")
plt.show()
```

```python
plt.hist(dataset['age'], bins = 5)
plt.show()
```

```python
sns.histplot(dataset['age'])
plt.show()
```

```python
sns.barplot(data=dataset, x='pclass', y='age')
plt.show()
```

```python
sns.barplot(data=dataset, x='pclass', y='age', hue = "sex")
plt.show()
```

```python
sns.boxplot(data = dataset, x = 'sex', y = 'age')
plt.show()
```

```python
sns.boxplot(data = dataset, x = 'sex', y = 'age', hue = "survived")
plt.show()
```

```python
sns.displot(dataset, x='age', hue='survived', kind='kde', fill=True, alpha=.5)
plt.show()
```

```python
pd.crosstab(dataset['pclass'], dataset['survived'])
```

```python
sns.heatmap(pd.crosstab(dataset['pclass'], dataset['survived']))
```

```python
sns.clustermap(pd.crosstab(dataset['parch'], dataset['survived']))
```
