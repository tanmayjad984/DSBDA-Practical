```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data= pd.read_csv("iris.csv")
plt.figure(figsize = (10,7))

x = data["SepalLengthCm"]
plt.hist(x, bins= 20 , color = "green")
plt.title("Sepal Length in cm")
plt.xlabel("SepalLengthCm")
plt.ylabel("Count")
```
